```python
# Importing required modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# to produce nice plots 
plt.style.use('fivethirtyeight')
%matplotlib inline
```

```python
# Read the data
stock_data=pd.read_csv("stock_data.csv",parse_dates=True,index_col="Date")
print(stock_data.shape)

stock_data=stock_data.dropna(axis=0)
stock_data.isnull().sum()
```

```python
# Read the data
benchmark_data=pd.read_csv("benchmark_data.csv",parse_dates=True,index_col="Date")
print(benchmark_data.shape)

benchmark_data=benchmark_data.dropna(axis=0)
benchmark_data.isnull().sum()
```

Task :- 5

```python
# calculate daily stock_data returns
stock_returns = stock_data.pct_change()

# plot daily returns
stock_returns.plot()
plt.show()
```

```python
# describe
stock_returns.describe()
```

Task :- 6

```python
# calculate daily benchmark_data returns
sp_returns = benchmark_data['S&P 500'].pct_change()

# plot daily returns
sp_returns.plot()
plt.show()
```

```python
# describe
sp_returns.describe()
```

Task :- 7

```python
# calculate difference in daily returns
excess_returns = stock_returns.sub(sp_returns, axis=0)

# plot excess_returns
excess_returns.plot()
plt.show()
```

```python
# describe
excess_returns.describe()
```

Task :- 8

```python
# calculate mean of excess_returns 
avg_excess_return = excess_returns.mean()

# plot avg_excess_returns
avg_excess_return.plot.bar()
plt_title = 'Mean of the Return Difference'
plt.show()
```

Task :- 9

```python
# calculate standard deviations
sd_excess_return = excess_returns.std()

# plot standard deviations
sd_excess_return.plot()
plt_title = 'Standard Deviation of the Return Difference'
plt.show()
```

Task :- 10

```python
# calculate daily sharpe ratio
daily_sharpe_ratio = avg_excess_return.div(sd_excess_return)

# annualize sharpe ratio
annual_factor = np.sqrt(252)
annual_sharpe_ratio = daily_sharpe_ratio.mul(annual_factor)

# plot annualized sharpe ratio
annual_sharpe_ratio.plot()
plt_title = 'Annualized Sharpe Ratio: Stocks vs S&P 500'
plt.show()
```
